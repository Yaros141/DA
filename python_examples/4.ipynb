```python
import pandas as pd
import numpy as np

vehicle_type = ['STANDARD', 'MINIVAN', 'VIP', 'MINIBUS']

capacity_map = {'STANDARD':'1-2',
                'MINIVAN':'1-3',
                'VIP':'1-3',
                'MINIBUS':'1-6'}
            

subregion_count = {'Dubai': 6,
                   'Abu Dhabi': 6,
                   'Fujairah': 3,
                   'Ras Al Khaimah': 3,
                   'Ajman': 2,
                   'Sharjah': 2,
                   'Jumeirah': 1,
                   'UAQ': 1,
                   'Palm Jumeirah': 1,
                   'Jebel Ali': 1}

output = pd.DataFrame(columns={'SERVICE NAME', 
                               'HOTEL', 
                               'MEAL PLAN', 
                               'DEPARTURE TOWN/REGION', 
                               'ARRIVAL TOWN/REGION',
                               'AIRLINE',
                               'TYPE OF TRANSPORT',
                               'FLIGHT',
                               'SEAT CLASS',
                               'ROOM TYPE',
                               'PAX COUNT',
                               'ADULT COUNT',
                               'CHILD COUNT',
                               'INFANT COUNT',
                               'TOUR NIGHTS',
                               'SERVICE NIGHTS',
                               'PER DAY/PER NIGHT',
                               'SERVICE DATES',
                               'RESERVATION DATES',
                               'NET PRICE',
                               'TO PRICE',
                               'SALE PRICE'})

input = pd.read_csv('transfer_input2.csv', delimiter = ';', encoding='windows-1251')
region_map = pd.read_csv('import.csv', delimiter = ';', encoding='windows-1251')


input_regions = input.Town.unique()

temp_df = pd.DataFrame(columns={'Town', 'Airport', 'Type', 'Price'})

for i in vehicle_type:
    
    for j in input_regions:
        count =input[input.Town == j]

        for k in range(0, (count.shape[0])):
            
            temp_search = count.iloc[k]
        
            temp_df = temp_df.append({'Town': temp_search.Town, 
                                      'Airport': temp_search.AIRPORT, 
                                      'Type' : i, 
                                      'Price':temp_search[i]}, ignore_index=True) 
            
temp_df = temp_df.dropna(how='any')

temp_df2 = pd.DataFrame(columns={'Town', 'Town2', 'Airport', 'Type', 'Price'})

for i in input_regions:
    count_map = region_map[region_map.Name1 == i]
    count_df = temp_df[temp_df.Town == i]
    
    for k in range(0,count_df.shape[0]):
        temp_map = count_df.iloc[k]
                
        for j in range(0, count_map.shape[0]):
            temp_search2 = count_map.iloc[j]

            temp_df2 = temp_df2.append({'Town2': temp_search2.Name2,
                                        'Town': temp_map.Town,
                                        'Airport': temp_map.Airport, 
                                        'Type' : temp_map.Type, 
                                        'Price':temp_map.Price}, ignore_index=True) 
        
temp_df2.Price = temp_df2.Price.astype(int)

for z in range(0, (temp_df2.shape[0])):
    
    output = output.append({'SERVICE NAME':'Transfer',
                            'HOTEL':'All',
                            'MEAL PLAN': ' All',
                            'DEPARTURE TOWN/REGION': temp_df2.Town2[z], 
                            'ARRIVAL TOWN/REGION': temp_df2.Airport[z],
                            'AIRLINE': 'All',
                            'TYPE OF TRANSPORT': temp_df2.Type[z],
                            'NET PRICE': temp_df2.Price[z]}, ignore_index=True)
    
output['PAX COUNT'] = output['TYPE OF TRANSPORT'].map(capacity_map)

output = output[['SERVICE NAME', 
               'HOTEL', 
               'MEAL PLAN', 
               'DEPARTURE TOWN/REGION', 
               'ARRIVAL TOWN/REGION',
               'AIRLINE',
               'TYPE OF TRANSPORT',
               'FLIGHT',
               'SEAT CLASS',
               'ROOM TYPE',
               'PAX COUNT',
               'ADULT COUNT',
               'CHILD COUNT',
               'INFANT COUNT',
               'TOUR NIGHTS',
               'SERVICE NIGHTS',
               'PER DAY/PER NIGHT',
               'SERVICE DATES',
               'RESERVATION DATES',
               'NET PRICE',
               'TO PRICE',
               'SALE PRICE']]

output.to_excel('output.xlsx')
```
